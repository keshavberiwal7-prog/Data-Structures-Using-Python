Q1. Swapping two elements in an array

```python
def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]
    return arr
arr = [1, 2, 3, 4]
print(swap(arr, 1, 3))
```

Q2. Splitting an array

```python
def split_array(arr):
    mid = len(arr) // 2
    return arr[:mid], arr[mid:]

arr = [1, 2, 3, 4, 5, 6]
a, b = split_array(arr)
print(a, b)
```

Q3. Calculator using User-Defined Functions

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b if b != 0 else "Error"

print(add(5, 3))
print(divide(10, 2))
```

Q4.Discount Calculator

```python
def calculate_discount(price, discount_percent):
    discount = (price * discount_percent) / 100
    return price - discount

print(calculate_discount(100, 20))
```

Q5.Grade Calculator

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "Fail"

print(calculate_grade(82))
```

Q6. Odd/Even Check

```python
def check_even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

print(check_even_odd(7))
```
